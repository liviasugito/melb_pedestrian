```python
# Import Libraries

import pandas as pd
```

# Pedestrian Volume Data

```python
# Read in the pedestrian volume data

ped_volume = pd.read_csv('../data/Pedestrian_Counting_System_-_Monthly__counts_per_hour_.csv')
```

```python
ped_volume.tail()
```

```python
ped_volume.dtypes
```

```python
ped_volume = ped_volume.drop("ID", axis=1)
```

```python
# No empty cell presents

ped_volume.isna().any(axis=None)
```

```python
# Extract the time
pattern = r'([0-9]*/)*(20[0-9]*) '
```

```python
ped_volume["Time"] = ped_volume["Date_Time"].str.replace(pattern, '', regex=True)
ped_volume = ped_volume.drop("Date_Time", axis=1)
```

```python
ped_volume
```

# Sensor Location Data

```python
# Read in the pedestrian sensor location data

sensor_loc = pd.read_csv('../data/Pedestrian_Counting_System_-_Sensor_Locations.csv')
sensor_loc
```

### Inactive or Removed Sensors

```python
sensor_loc[sensor_loc["status"] != "A"]
```

```python
sensor_loc = sensor_loc.drop(["sensor_name", "location", "note", "sensor_description"], axis=1)
```

Four of them has empty directions. These instances won't be dropped for now as they have substantial pedestrian count in the other dataframe.

```python
# Join the two dataframe
complete_df = ped_volume.merge(sensor_loc, left_on='Sensor_ID', right_on='sensor_id')
```
